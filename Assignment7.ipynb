b)	Find out top 10 cities in total number of graduates
c)	Find out top 10 cities and their locations in respect of  total effective_literacy_rate.



```python
df=pd.read_csv(r"C:\Users\Anuradha\Desktop\InClass_DataPreprocessing_datasets\Data sets\datasets\Indian_cities.csv")
```

```python
df.head()
```

```python
for col in df.columns:
    print(col)
```

a)	Find out top 10 states in female-male sex ratio


```python
sex_ratio = df.groupby('state_name').agg({'sex_ratio': 'mean'}).sort_values(by='sex_ratio', ascending=False).head(10)
print("Top 10 states in terms of female-male sex ratio:")
print(sex_ratio)
```


b)	Find out top 10 cities in total number of graduates


```python
graduate=df.groupby('name_of_city').agg({'total_graduates':'mean'}).sort_values(by='total_graduates',ascending=False).head(10)
print(graduate)
```


c)	Find out top 10 cities and their locations in respect of  total effective_literacy_rate.

```python
cities_literacy = df.nlargest(10, 'effective_literacy_rate_total')
cities_literacy = cities_literacy[['name_of_city', 'location', 'effective_literacy_rate_total']]

print("Top 10 cities in terms of total effective literacy rate:")
print(cities_literacy)
```
